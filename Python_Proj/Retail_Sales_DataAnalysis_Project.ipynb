```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Once you’ve loaded the data inside your class as shown, it’s stored in the instance variable self.data.
#To read or access the data, you just need to create an object (instance) of your class, then refer to that variable.
```

```python
class RetailSalesAnalyzer:
  def __init__(self):
    self.data = pd.read_csv('/content/retail_sales.csv')
    self.data['Date'] = pd.to_datetime(self.data['Date'])

  # Cleaning Data
  def clean_data(self):
    self.data.dropna(inplace=True)

  # Data Manipulation and Analysis:
  # Calculate total sales per product.
  def total_sales_per_product(self):
    return self.data.groupby('Product')['Sales'].sum()

  # Identify the best-selling product.
  def best_selling_product(self):
    return self.total_sales_per_product().idxmax()

  # Compute average daily sales.
  def average_daily_sales(self):
    return self.data['Sales'].mean()


  # Visualization
  def plot_sales_trend(self):
    self.data.groupby('Date')['Sales'].sum().plot(kind='line')
    plt.title('Sales Trend Over Time')
    plt.xlabel('Date')
    plt.ylabel('Total Sales')
    plt.show()

  # Sales per product in bar chart
  def plot_sales_per_product(self):
    sales_per_product = self.total_sales_per_product()
    sales_per_product.plot(kind='bar')
    plt.title('sales per product')
    plt.xlabel('Product')
    plt.ylabel('Total Sales')
    plt.show()




  # 5.Class and Function, script ececution
  analyzer = RetailSalesAnalyzer()
  analyzer.clean_data()
  print(analyzer.total_sales_per_product())
  print(analyzer.best_selling_product())
  print(analyzer.average_daily_sales())
  analyzer.plot_sales_trend()
  analyzer.plot_sales_per_product()






```
